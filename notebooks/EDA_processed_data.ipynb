```python
import pandas as pd
from sqlalchemy import create_engine
from dotenv import load_dotenv
import os

load_dotenv()

DB_HOST = os.getenv('DB_HOST')
DB_PORT = os.getenv('DB_PORT')
DB_NAME = os.getenv('DB_NAME')
DB_USER = os.getenv('DB_USER')
DB_PASS = os.getenv('DB_PASS')
DB_SCHEMA = os.getenv('DB_SCHEMA')

DATABASE_URL = f"postgresql://{DB_USER}:{DB_PASS}@{DB_HOST}:{DB_PORT}/{DB_NAME}"

engine = create_engine(DATABASE_URL)
```

```python
df = pd.read_sql("SELECT * FROM stg_one_hot_encoding", engine)
df
```

```python
df.info()
```

```python
df.columns
```

```python
# Tratamento das datas
df['data_nascimento'] = pd.to_datetime(df['data_nascimento'], errors='coerce')
df['data_cadastro'] = pd.to_datetime(df['data_cadastro'], errors='coerce')
df['idade'] = (pd.to_datetime('today') - df['data_nascimento']).dt.days // 365
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 5))
sns.countplot(data=df, x='sexo', hue='sexo', palette='coolwarm', legend=False)
plt.title("Distribuição por Sexo")
plt.show()
```

```python
plt.figure(figsize=(12, 5))
sns.histplot(df['idade'], bins=30, kde=True)
plt.title("Distribuição de Idade")
plt.show()
# Maior parte dos pacientes tem entre 20 e 40 anos
```

```python
plt.figure(figsize=(12, 5))
sns.histplot(df[df['obito'] == 1]['idade'], bins=30, kde=True, color='red')
plt.title("Idade dos Pacientes que Foram a Óbito")
plt.show()
```

A alta incidência de óbitos entre indivíduos de 20 a 40 anos sugere que esses falecimentos não seguem o padrão natural de envelhecimento e declínio fisiológico, caracterizando-se, portanto, como mortes precoces. Esse fenômeno pode estar associado a diversos fatores externos e condições de saúde que afetam de forma desproporcional essa faixa etária.

```python
plt.figure(figsize=(12, 5))
sns.histplot(df[df['obito'] == 1]['renda_familiar'], kde=False, color='red')
plt.xticks(rotation=45, ha="right")
plt.title("Renda Familiar dos Pacientes que Foram a Óbito")
plt.show()
```

Os dados indicam que a maioria dos pacientes que foram a óbito possuía uma renda entre 1 e 2 salários mínimos, sugerindo que fatores socioeconômicos desempenham um papel central na mortalidade dessa população

```python
plt.figure(figsize=(10, 5))
df_obitos = df[df['obito'] == 1]
df_obitos["bairro"].value_counts().head(10).plot(kind="bar", color="crimson")
plt.title("Top 10 Bairros com Mais Óbitos")
plt.xlabel("Bairro")
plt.ylabel("Óbitos")
plt.xticks(rotation=45)
plt.show()
```

Os 10 bairros com mais óbitos podem ser caracterizados, em geral, por uma combinação de fatores socioeconômicos e logísticos que influenciam diretamente a qualidade de vida de seus moradores e contribuem para esse cenário alarmante. Muitos desses bairros possuem condições de renda mais baixas, o que pode resultar em um acesso limitado a serviços de saúde de qualidade e a outros recursos essenciais para a promoção de uma vida saudável.

```python
plt.figure(figsize=(12, 5))
sns.boxplot(data=df, y='n_atendimentos_atencao_primaria')
plt.title("Distribuição de Atendimentos na Atenção Primária")
plt.show()
```

```python
plt.figure(figsize=(12, 5))
sns.boxplot(data=df, y='n_atendimentos_hospital')
plt.title("Distribuição de Atendimentos em Hospital")
plt.show()
```

```python
transportes = ['onibus', 'metro', 'carro', 'bicicleta', 'caminhao']
freq_atendimento = {t: df.loc[df[t] == 1, 'n_atendimentos_atencao_primaria'].sum() for t in transportes}

plt.figure(figsize=(8, 5))
plt.bar(freq_atendimento.keys(), freq_atendimento.values(), color=['blue', 'green', 'red'])

plt.xlabel('Meio de Transporte')
plt.ylabel('Total de Atendimentos')
plt.title('Frequência de Atendimentos por Meio de Transporte')
plt.xticks(rotation=30)  

plt.show()
```

A maior frequência de atendimentos a pacientes que dependem de ônibus e transporte público é um indicador de uma série de desigualdades estruturais que esses indivíduos enfrentam, especialmente no que diz respeito ao acesso a serviços de saúde e outras condições essenciais. O fato de dependerem do transporte público para se deslocar até unidades de saúde, hospitais e outros serviços significa que eles estão, muitas vezes, expostos a barreiras geográficas, econômicas e logísticas que dificultam ou até impossibilitam o acesso adequado aos cuidados necessários.

```python
transportes = ['onibus', 'metro', 'carro', 'bicicleta', 'caminhao']
freq_atendimento = {t: df.loc[df[t] == 1, 'n_atendimentos_hospital'].sum() for t in transportes}

plt.figure(figsize=(8, 5))
plt.bar(freq_atendimento.keys(), freq_atendimento.values(), color=['blue', 'green', 'red'])

plt.xlabel('Meio de Transporte')
plt.ylabel('Total de Atendimentos')
plt.title('Frequência de Atendimentos por Meio de Transporte')
plt.xticks(rotation=30)  

plt.show()
```

Os dados evidenciam que as condições precárias de transporte e as limitações financeiras não apenas aumentam a barreira física para o acesso aos serviços de saúde, mas também refletem uma série de desigualdades sociais que vão além do setor de saúde em si. Para melhorar o acesso dessas populações a cuidados adequados, é fundamental que haja uma abordagem mais ampla, que envolva melhorias não só no sistema de saúde, mas também nas condições de transporte, educação, segurança e na redução das desigualdades econômicas.
